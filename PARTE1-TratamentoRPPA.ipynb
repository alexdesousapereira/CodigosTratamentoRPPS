## Parte 02 - Aposentados

```python
aposentados = pd.read_excel("Dados.xlsx",sheet_name = "APOSENTADOS")
```

### VL_APOSENTADORIA

```python
# Valida valores com base em lista passada
c = 0
indexerrados = []
for i in range(len(aposentados)):
    if aposentados['VL_APOSENTADORIA'][i] < 1212:
            indexerrados.append(c)            
    c = c + 1
if len(indexerrados) > 0:
    print('Existem ',len(indexerrados),' argumentos errados')
```

```python
dados = aposentados.drop(index=indexerrados)
media = dados[['CO_TIPO_CARGO','CO_TIPO_POPULACAO','CO_TIPO_APOSENTADORIA',
    'VL_APOSENTADORIA']].groupby(
        by=['CO_TIPO_CARGO','CO_TIPO_POPULACAO','CO_TIPO_APOSENTADORIA'],
        as_index=False).mean()
for i in indexerrados:
    valor = media.loc[(media['CO_TIPO_CARGO'] == aposentados['CO_TIPO_CARGO'][i])
     & (media['CO_TIPO_POPULACAO'] == aposentados['CO_TIPO_POPULACAO'][i]) &
    (media['CO_TIPO_APOSENTADORIA'] == aposentados['CO_TIPO_APOSENTADORIA'][i]),
        ['VL_APOSENTADORIA']]
    aposentados['VL_APOSENTADORIA'][i] = valor['VL_APOSENTADORIA']
```

### VL_CONTRIBUICAO

```python
# Valida valores com base em lista passada
c = 0
indexerrados = []
for i in range(len(aposentados)):
    if aposentados['VL_APOSENTADORIA'][i] > aposentados['VL_TETO_ESPECIFICO'][i]:
        if aposentados['VL_CONTRIBUICAO'][i] != (aposentados['VL_TETO_ESPECIFICO'][i] - aposentados['VL_APOSENTADORIA'][i] ) * 0.14:
            indexerrados.append(c)
    elif aposentados['VL_CONTRIBUICAO'][i] != 0:
        indexerrados.append(c)
    c = c + 1
if len(indexerrados) > 0:
    print('Existem ',len(indexerrados),' argumentos errados')
```

```python
for i in indexerrados:
    if aposentados['VL_APOSENTADORIA'][i] > aposentados['VL_TETO_ESPECIFICO'][i]:
        aposentados['VL_CONTRIBUICAO'][i] = (aposentados['VL_TETO_ESPECIFICO'][i] - aposentados['VL_APOSENTADORIA'][i] ) * 0.14
    else:
         aposentados['VL_CONTRIBUICAO'][i] = 0
```

### Gerando novo conjunto de dados 


```python
aposentados.to_csv('APOSENTADOS.csv',sep=";")
```

## Parte 03 - Dependentes

```python
dependentes = pd.read_excel("Dados.xlsx",sheet_name = "DEPENDENTES")
```

### idades_dependentes

```python
# Valida valores com base em lista passada
c = 0
indexerrados = []
for i in range(len(dependentes)):
    if dependentes['CO_TIPO_DEPENDECIA'][i] not in [1,2,3,4,5,6]:
        indexerrados.append(c)
    elif dependentes['CO_TIPO_DEPENDECIA'][i]  == 1 and dependentes['idades_dependentes'][i] < 18:
        indexerrados.append(c)
    elif dependentes['CO_TIPO_DEPENDECIA'][i]  == 2 and dependentes['idades_dependentes'][i] >= 21:
        indexerrados.append(c)
    elif dependentes['CO_TIPO_DEPENDECIA'][i]  == 4 and dependentes['idades_dependentes'][i] < 18:
        indexerrados.append(c)
    elif dependentes['CO_TIPO_DEPENDECIA'][i]  == 5 and dependentes['idades_dependentes'][i] >= 21:
        indexerrados.append(c)
    c = c + 1
if len(indexerrados) > 0:
    print('Existem ',len(indexerrados),' argumentos errados')
```

```python
dados = dependentes.drop(index=indexerrados)
media = dados[['CO_TIPO_DEPENDECIA','idades_dependentes']].groupby(by=['CO_TIPO_DEPENDECIA'],as_index=False).mean()
for i in range(len(dependentes)):
    if dependentes['CO_TIPO_DEPENDECIA'][i]  == 1 and dependentes['idades_dependentes'][i] < 18:
        valor = media.loc[(media['CO_TIPO_DEPENDECIA'] == 1),['idades_dependentes']]
        dependentes['idades_dependentes'][i] = valor['idades_dependentes']
    elif dependentes['CO_TIPO_DEPENDECIA'][i]  == 2 and dependentes['idades_dependentes'][i] >= 21:
        valor = media.loc[(media['CO_TIPO_DEPENDECIA'] == 2),['idades_dependentes']]
        dependentes['idades_dependentes'][i] = valor['idades_dependentes']
    elif dependentes['CO_TIPO_DEPENDECIA'][i]  == 4 and dependentes['idades_dependentes'][i] >= 18:
        valor = media.loc[(media['CO_TIPO_DEPENDECIA'] == 4),['idades_dependentes']]
        dependentes['idades_dependentes'][i] = valor['idades_dependentes']
    elif dependentes['CO_TIPO_DEPENDECIA'][i]  == 5 and dependentes['idades_dependentes'][i] >= 21:
        valor = media.loc[(media['CO_TIPO_DEPENDECIA'] == 5),['idades_dependentes']]
        dependentes['idades_dependentes'][i] = valor['idades_dependentes']
```

```python
dependentes.to_csv('DEPENDENTES.csv',sep=";")
```

## Parte 04 - Pensões

```python
pensionistas = pd.read_excel("Dados.xlsx",sheet_name = "PENSIONISTAS")
```

### VL_BENEF_PENSAO

```python
# Valida valores com base em lista passada
c = 0
indexerrados = []
for i in range(len(pensionistas)):
    if pensionistas['VL_BENEF_PENSAO'][i] < 1212:
        indexerrados.append(c)            
    c = c + 1
if len(indexerrados) > 0:
    print('Existem ',len(indexerrados),' argumentos errados')
```

```python
dados = pensionistas.drop(index=indexerrados)
media = dados[['CO_SEXO_INSTITUIDOR','CO_TIPO_RELACAO','VL_BENEF_PENSAO']].groupby(by=['CO_SEXO_INSTITUIDOR','CO_TIPO_RELACAO'],as_index=False).mean()
for i in indexerrados:
    valor = media.loc[(media['CO_SEXO_INSTITUIDOR'] == pensionistas['CO_SEXO_INSTITUIDOR'][i]) & (media['CO_TIPO_RELACAO'] == pensionistas['CO_TIPO_RELACAO'][i]),['VL_BENEF_PENSAO']]
    pensionistas['VL_BENEF_PENSAO'][i] = valor['VL_BENEF_PENSAO']
```

### VL_CONTRIBUICAO

```python
# Valida valores com base em lista passada
c = 0
teto = 7087.22
indexerrados = []
for i in range(len(pensionistas)):
    if pensionistas['VL_BENEF_PENSAO'][i] > teto:
        if pensionistas['VL_CONTRIBUICAO'][i] != (teto - pensionistas['VL_BENEF_PENSAO'][i] ) * 0.14:
            indexerrados.append(c)
    elif pensionistas['VL_CONTRIBUICAO'][i] != 0:
        indexerrados.append(c)
    c = c + 1
if len(indexerrados) > 0:
    print('Existem ',len(indexerrados),' argumentos errados')
```

```python
for i in indexerrados:
    if pensionistas['VL_BENEF_PENSAO'][i] > teto:
         pensionistas['VL_BENEF_PENSAO'][i]  = (teto - pensionistas['VL_BENEF_PENSAO'][i] ) * 0.14
    else:
         pensionistas['VL_CONTRIBUICAO'][i] = 0
```

### Gerando novo conjunto de dados 


```python
pensionistas.to_csv('Planilhas-Tratadas/PENSIONISTAS.csv',sep=";")
```
